```python
from sklearn.model_selection import train_test_split
```

```python
x=housig_df.drop(['SalePrice'],axis=1)
y=housig_df.SalePrice
```

```python
x_train,x_test,y_train,y_test=train_test_split(housig_df.drop(labels=['SalePrice'],axis=1),
                                                              housig_df['SalePrice'],
                                                              test_size=0.3,random_state=42)
```

```python
x_train.head()
```

```python
x_train.isnull().sum()
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
mutual=mutual_info_regression(x_train.fillna(0),y_train)
```

```python
mutual
```

```python
mutual=pd.Series(mutual)
mutual.index=x_train.columns
mutual.sort_values(ascending=False)
```

```python
mutual.sort_values(ascending=False).plot.bar(figsize=(15,7))
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
selc_five_col=SelectKBest(mutual,k=5)
selc_five_col.fit(x_train,y_train)
x_train.columns[selc_five_col.get_support()]
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
selc=SelectPercentile(mutual,percentile=20)
selc.fit(x_train.fillna(0),y_train)
x_train.columns[selc.ge]
```
